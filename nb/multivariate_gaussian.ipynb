```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000

# uncorrelated
x = np.random.randn(N) # mu + signa*np.random.randn(N)
y = np.random.randn(N)
plt.scatter(x,y)
plt.xlim(-6., 6.)
plt.ylim(-6., 6.)
plt.show()
print(np.cov(x,y))

# sigma_y = 2*sigma_x
x = np.random.randn(N)
y = 2.*np.random.randn(N)
plt.scatter(x,y)
plt.xlim(-6., 6.)
plt.ylim(-6., 6.)
plt.show()
print(np.cov(x,y))

# covariance
x = np.random.randn(N)
y = x + np.random.randn(N)
plt.scatter(x,y)
plt.xlim(-6., 6.)
plt.ylim(-6., 6.)
plt.show()
print(np.cov(x,y))
```
