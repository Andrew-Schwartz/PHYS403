```python
import numpy as np
import matplotlib.pyplot as plt

def f_linear(x):
    return x
def f_sqrt(x):
    return np.sqrt(x)
def f_quad(x):
    return x**2

sigma = 0.1
N = 11

# Generate data with different models
x = [0.1*i for i in range(N)]
y_linear = [] # linear
y_sqrt = [] # square root
y_quad = [] # quadratic
y_err = [sigma for i in range(N)]
for i in range(N):
    y_linear.append(np.random.normal(x[i], sigma))
    y_sqrt.append(np.random.normal(np.sqrt(x[i]), sigma))
    y_quad.append(np.random.normal(x[i]**2, sigma))
    
# make functions of the models for plotting
xc = np.linspace(0, 1, 1000)
fl = f_linear(xc)
fs = f_sqrt(xc)
fq = f_quad(xc)

# determine the log-likelihood of each data, given the linear model
ll_linear = 0.
ll_quad = 0.
ll_sqrt = 0.
for i in range(N):
    ll_linear -= (y_linear[i]-x[i])**2/sigma**2
    ll_quad -= (y_quad[i]-x[i])**2/sigma**2
    ll_sqrt -= (y_sqrt[i]-x[i])**2/sigma**2
    
print(ll_linear, ll_quad, ll_sqrt)

plt.scatter(x, y_linear)
plt.errorbar(x, y_linear, yerr=y_err, fmt='o')
plt.plot(xc,fl)
plt.plot(xc,fs)
plt.plot(xc,fq)
plt.show()
```
